```python


import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

#Data visualization
import matplotlib.pyplot as plt
import seaborn as sns

```

### Reading the csv file data

```python
df = pd.read_csv('menu.csv')
df.isnull().sum()
```

```python
df.head(5)
```

 ### Okay data looks cleaner than what I have expected and from the dataframe head, it can be inferred that the data  consist of a column "item" which contains food item names and other columns describe its nutrition value

*   ### *Vitamin A* Swarmplot

```python
plot = sns.swarmplot(x="Category", y="Vitamin A (% Daily Value)", data=df)
plt.setp(plot.get_xticklabels(),rotation=70)
plt.title("Vitamin A (% Daily Value)")
plt.show()
```

Well, it's a known fact that salads(vegetable) contain vitamin A in abundant amount. Let's find out for which items we are getting maximum and minimum value for vitamin A

```python
df.at[df["Vitamin A (% Daily Value)"].idxmax(),'Item']
```

```python
df.at[df["Vitamin A (% Daily Value)"].idxmin(),'Item']
```

so, its 'Premium Bacon Ranch Salad(without Chicken)' and 'Sausage Biscuit (Regular Biscuit)' respectively.

```python
df_numeric = df.select_dtypes(include=['float64', 'int64'])

# Calculate correlation matrix
corr = df_numeric.corr()

# Plot the correlation heatmap
plt.figure(figsize=(10, 10))
sns.heatmap(corr, mask=np.zeros_like(corr, dtype=bool), cmap=sns.diverging_palette(220, 10, as_cmap=True),
            square=True)
plt.title("McD Correlation Heatmap")
plt.show()
```

Well the correlation matrix confirms the finding that we have discovered above. 

## Conclusion

### What to avoid?
*  Chicken McNuggets:- Although its provide you ample amount of protein but its also contain huge amount of calories (more than 1750).
*  Big Breakfast with Hotcakes (Regular Biscuit):- contain excess amount of cholesterol.
*  Double Quarter Pounder with Cheese :- If you are on diet than avoid this item, it contain a bumper amount of trans fat

### Safe options
* Hash Brown
* Diet Coke (Small)
